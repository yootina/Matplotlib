# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

# autopct : 부채꼴 안에 표시될 숫자의 형식을 지정(소수점 한자리까지 표시하도록 설정)
# startangle : 부채꼴이 그려지는 시작 각도를 설정(디폴트는 0도로 설정, counterclock=False로 설정하면 시계 방향 순서로 표시)

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

# explode : 부채꼴이 파이 차트의 중심에서 벗어나는 정도를 설정
# explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.05] * 6

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)

# 범례가 원그래프를 가릴 때 x값과 y값 설정 후 범례 title 설정
plt.legend(loc=(1.2, 0.3), title='언어별 선호도')
plt.show()
```
