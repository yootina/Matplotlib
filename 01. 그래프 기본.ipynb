# Matplotlib

- 선 그래프(line plot) : plot()

- 막대그래프 (bar chart) : bar()
    - 가로 막대그래프 (barh plot) : barh()

- 산점도 (scatter plot) : scatter()

- 히스토그램 (histogram) : hist()

- 상자 그림 (box plot) : boxplot()

- 파이 차트 (pie chart) : pie()

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x,y)
```

```python
print(plt.plot(x,y))
```

```python
plt.plot(x,y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist] # 위의 코드로 입력하게되면 상세정보까지 다 나오니깐 줄여서 확인하는 방법
```
