# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=5)
```

## 마커(marker)

```python
plt.plot(x, y, marker='o')
```

```python
plt.plot(x, y, marker='o', linestyle='None')
```

```python
plt.plot(x, y, marker='v')
```

```python
plt.plot(x, y, marker='v', markersize = 10)
```

```python
plt.plot(x, y, marker='x', markersize = 10)
```

```python
plt.plot(x, y, marker='X', markersize = 10, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize = 10, markeredgecolor='red')
```

```python
plt.plot(x, y, marker='o', markersize = 10, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일
[공식문서 링크](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python
plt.plot(x, y, linestyle=':')
```

```python
plt.plot(x, y, linestyle='--')
```

```python
plt.plot(x, y, linestyle='-.')
```

```python
plt.plot(x, y, linestyle='-')
```

## 색깔
[공식문서 링크](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
plt.plot(x, y, color='pink')
```

```python
plt.plot(x, y, color='#ff0000')
```

```python
plt.plot(x, y, color='b')
```

```python
plt.plot(x, y, color='g')
```

```python
plt.plot(x, y, color='#00ffff')
```

## 포맷

```python
plt.plot(x, y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x, y, 'bv:') # color, marker, linestyle
```

```python
plt.plot(x, y, 'go') # color, marker, linestyle = None 
```

```python
plt.plot(x, y, color='g', marker='o', linestyle='None') # 위 코드와 동일한 코드
```

## 축약어
[공식문서 링크](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
plt.plot(x, y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

## 투명도

```python
# alpha: 투명도 (0~1)
plt.plot(x, y, marker='o', mfc='red', ms=10, alpha=0.3)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x, y)
```

```python
plt.figure(figsize=(5, 10))
plt.plot(x, y)
```

```python
# dots per inch, 확대
plt.figure(figsize=(10, 5), dpi=200)
plt.plot(x, y)
```

## 배경색 지정

```python
plt.figure(facecolor='yellow')
plt.plot(x, y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x, y)
```
