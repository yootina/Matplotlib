# 15. 산점도 그래프

산점도 (Scatter plot)는 두 변수의 상관 관계를 직교 좌표계의 평면에 점으로 표현하는 그래프입니다.

`matplotlib.pyplot` 모듈의 `scatter()` 함수를 이용하면 산점도를 그릴 수 있습니다.

**Keyword: plt.scatter(), scatter plot, 산점도**

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_csv('../Pandas/score.csv')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

#### 기본 사용 방법


```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

NumPy의 random 모듈에 포함된 rand() 함수를 사용해서 8개의 랜덤값을 생성하고,

x, y 데이터를 순서대로 scatter() 함수에 입력하면 x, y 값에 해당하는 위치에 기본 마커가 표시됩니다.

```python
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
sizes = df['학년'] * 500 # 1학년=500, 2학년=1000, 3학년=1500
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
# cmap 파라미터에 컬러맵에 해당하는 문자열을 지정
# alpha 파라미터는 마커의 투명도를 지정(0~1사이)

plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='magma', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
```

```python
plt.figure(figsize=(7, 7))

plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='magma', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')

# orientation 컬러맵 가로로 정렬
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal')
plt.show()
```
