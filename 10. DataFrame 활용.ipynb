```python
plt.plot(df['지원번호'], df['영어'])
plt.plot(df['지원번호'], df['수학'])

plt.grid(axis='y') # y축에 대해서만 격자 생성
plt.show()
```

```python
plt.plot(df['지원번호'], df['영어'])
plt.plot(df['지원번호'], df['수학'])

plt.grid(axis='y', color='purple', alpha=0.5, linestyle='--', linewidth=2) # 컬러, 투명도, 라인스타일, 라인두께
plt.show()
```
