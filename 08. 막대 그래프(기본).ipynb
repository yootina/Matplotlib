## 08. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
labels = ['꼬부기', '파이리', '이상해씨'] # 이름
values = [250, 249, 253] # 능력치

plt.bar(labels, values)
```

```python
labels = ['꼬부기', '파이리', '이상해씨'] # 이름
values = [250, 249, 253] # 능력치

plt.bar(labels, values, color='r')
```

```python
labels = ['꼬부기', '파이리', '이상해씨'] # 이름
values = [250, 249, 253] # 능력치
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5) # alpha: 투명도(0~1사이)
```

```python
# 포켓몬들의 능력치를 차이를 도드라지게 보고싶을 때 ylim값을 주기

labels = ['꼬부기', '파이리', '이상해씨']
values = [250, 249, 253]

plt.bar(labels, values)
plt.ylim(240, 255)
```

```python
# width bar두께
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)

# x축의 이름 데이터 각도를 45도로 설정
plt.xticks(rotation=45)
# y축의 능력치 데이터 각도를 45도로 설정
plt.yticks(rotation=45)
```

```python
# 데이터는 똑같이 사용하지만 xtics 데이터만 바꾸고 싶을 때

labels = ['꼬부기', '파이리', '이상해씨']
values = [250, 249, 253]
ticks = ['번호 007', '번호 004', '번호 001']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=30)
```
