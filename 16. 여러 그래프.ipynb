# 16. 여러 그래프

matplotlib.pyplot 모듈의 `subplot()` 함수는 여러 개의 그래프를 하나의 그림에 나타내도록 합니다.

`subplot()` 함수를 사용해서 여러 개의 그래프를 나타내고, 축을 공유하는 방법을 소개합니다.

**Keyword: plt.subplot(), 여러 개의 그래프, 축 공유**

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_csv('../Pandas/score.csv')
df
```

```python
# 2x2에 해당하는 plot들을 생성
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('여러 그래프 넣기')

# axs[0,0] 위치(첫 번째 그래프)에 bar그래프를 그리기
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수') # 데이터 설정
axs[0, 0].set_title('첫 번째 그래프') # 제목
axs[0, 0].legend() # 범례 설정
axs[0, 0].set(xlabel='이름', ylabel='점수') # x, y축 라벨 설정
axs[0, 0].set_facecolor('lightyellow') # 전경색 수정
axs[0, 0].grid(ls='--', lw=0.5)

# 두 번째 그래프(axs[0, 1])
axs[0, 1].plot(df['이름'], df['수학'],label='수학점수')
axs[0, 1].plot(df['이름'], df['영어'],label='영어점수')
axs[0, 1].legend()

# 세 번째 그래프(axs[1, 0])
axs[1, 0].barh(df['이름'], df['키'])

# 네 번째 그래프(axs[1, 1])
axs[1, 1].plot(df['이름'], df['사회'],label='사회점수', color='pink', alpha=0.7)
axs[1, 1].legend()

plt.show()
```
